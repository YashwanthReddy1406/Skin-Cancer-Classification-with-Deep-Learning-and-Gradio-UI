```python
# Skin Cancer Classification with Deep Learning and Gradio UI
# Complete Google Colab Project

# Step 1: Install required packages
!pip install kagglehub gradio tensorflow pillow pandas numpy matplotlib seaborn scikit-learn -q

# Step 2: Import libraries
import kagglehub
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
import os
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import EfficientNetB0
from sklearn.model_selection import train_test_split
from sklearn.utils import class_weight
import gradio as gr
import warnings
warnings.filterwarnings('ignore')

# Step 3: Download and load dataset
print("Downloading dataset...")
path = kagglehub.dataset_download("kmader/skin-cancer-mnist-ham10000")
print("Path to dataset files:", path)

# Load metadata
metadata_path = os.path.join(path, "HAM10000_metadata.csv")
df = pd.read_csv(metadata_path)

# Image paths
img_dir1 = os.path.join(path, "HAM10000_images_part_1")
img_dir2 = os.path.join(path, "HAM10000_images_part_2")

print(f"\nDataset shape: {df.shape}")
print(f"\nClass distribution:\n{df['dx'].value_counts()}")

# Step 4: Data preprocessing
# Map diagnosis codes to full names
lesion_type_dict = {
    'nv': 'Melanocytic nevi',
    'mel': 'Melanoma',
    'bkl': 'Benign keratosis',
    'bcc': 'Basal cell carcinoma',
    'akiec': 'Actinic keratoses',
    'vasc': 'Vascular lesions',
    'df': 'Dermatofibroma'
}

df['cell_type'] = df['dx'].map(lesion_type_dict)
df['image_path'] = df['image_id'].apply(
    lambda x: os.path.join(img_dir1, x + '.jpg') if os.path.exists(os.path.join(img_dir1, x + '.jpg'))
    else os.path.join(img_dir2, x + '.jpg')
)

# Create label encoding
label_encoder = {label: idx for idx, label in enumerate(df['dx'].unique())}
df['label'] = df['dx'].map(label_encoder)

print("\nLabel encoding:", label_encoder)

# Step 5: Visualize data distribution
fig, ax = plt.subplots(figsize=(12, 6))
df['cell_type'].value_counts().plot(kind='bar', ax=ax, color='skyblue')
ax.set_title('Distribution of Skin Lesion Types', fontsize=16, fontweight='bold')
ax.set_xlabel('Lesion Type', fontsize=12)
ax.set_ylabel('Count', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Step 6: Display sample images
fig, axes = plt.subplots(2, 4, figsize=(16, 8))
axes = axes.ravel()
for i, lesion in enumerate(df['dx'].unique()):
    sample_img_path = df[df['dx'] == lesion]['image_path'].iloc[0]
    img = Image.open(sample_img_path)
    axes[i].imshow(img)
    axes[i].set_title(f"{lesion_type_dict[lesion]}", fontsize=12, fontweight='bold')
    axes[i].axis('off')
plt.tight_layout()
plt.show()

# Step 7: Split data
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42, stratify=df['label'])
train_df, val_df = train_test_split(train_df, test_size=0.15, random_state=42, stratify=train_df['label'])

print(f"\nTrain set: {len(train_df)}")
print(f"Validation set: {len(val_df)}")
print(f"Test set: {len(test_df)}")

# Step 8: Data augmentation and generators
IMG_SIZE = 224
BATCH_SIZE = 32

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True,
    vertical_flip=True,
    zoom_range=0.2,
    fill_mode='nearest'
)

val_test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_dataframe(
    train_df,
    x_col='image_path',
    y_col='dx',
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=True
)

val_generator = val_test_datagen.flow_from_dataframe(
    val_df,
    x_col='image_path',
    y_col='dx',
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

test_generator = val_test_datagen.flow_from_dataframe(
    test_df,
    x_col='image_path',
    y_col='dx',
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

# Step 9: Handle class imbalance
class_weights = class_weight.compute_class_weight(
    'balanced',
    classes=np.unique(train_df['label']),
    y=train_df['label']
)
class_weights_dict = dict(enumerate(class_weights))
print(f"\nClass weights: {class_weights_dict}")

# Step 10: Build model with transfer learning
def create_model():
    base_model = EfficientNetB0(
        include_top=False,
        weights='imagenet',
        input_shape=(IMG_SIZE, IMG_SIZE, 3)
    )

    # Freeze base model
    base_model.trainable = False

    # Create model
    model = keras.Sequential([
        base_model,
        layers.GlobalAveragePooling2D(),
        layers.BatchNormalization(),
        layers.Dropout(0.5),
        layers.Dense(256, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.3),
        layers.Dense(7, activation='softmax')
    ])

    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=0.001),
        loss='categorical_crossentropy',
        metrics=['accuracy', keras.metrics.AUC(name='auc')]
    )

    return model

model = create_model()
model.summary()

# Step 11: Train model
print("\n" + "="*50)
print("Starting model training...")
print("="*50)

callbacks = [
    keras.callbacks.EarlyStopping(
        monitor='val_loss',
        patience=5,
        restore_best_weights=True,
        verbose=1
    ),
    keras.callbacks.ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=3,
        min_lr=1e-7,
        verbose=1
    )
]

history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=20,
    class_weight=class_weights_dict,
    callbacks=callbacks,
    verbose=1
)

# Step 12: Evaluate model
print("\n" + "="*50)
print("Evaluating model on test set...")
print("="*50)

test_loss, test_accuracy, test_auc = model.evaluate(test_generator)
print(f"\nTest Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f}")
print(f"Test AUC: {test_auc:.4f}")

# Step 13: Plot training history
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Accuracy
axes[0].plot(history.history['accuracy'], label='Train Accuracy', linewidth=2)
axes[0].plot(history.history['val_accuracy'], label='Val Accuracy', linewidth=2)
axes[0].set_title('Model Accuracy', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Accuracy')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Loss
axes[1].plot(history.history['loss'], label='Train Loss', linewidth=2)
axes[1].plot(history.history['val_loss'], label='Val Loss', linewidth=2)
axes[1].set_title('Model Loss', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Loss')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

# AUC
axes[2].plot(history.history['auc'], label='Train AUC', linewidth=2)
axes[2].plot(history.history['val_auc'], label='Val AUC', linewidth=2)
axes[2].set_title('Model AUC', fontsize=14, fontweight='bold')
axes[2].set_xlabel('Epoch')
axes[2].set_ylabel('AUC')
axes[2].legend()
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Step 14: Save model
model.save('skin_cancer_model.h5')
print("\nModel saved as 'skin_cancer_model.h5'")

# Step 15: Create Gradio UI
def predict_skin_lesion(image):
    """Predict skin lesion type from image"""
    if image is None:
        return "Please upload an image"

    # Preprocess image
    img = Image.fromarray(image.astype('uint8'), 'RGB')
    img = img.resize((IMG_SIZE, IMG_SIZE))
    img_array = np.array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)

    # Predict
    predictions = model.predict(img_array, verbose=0)
    predicted_class = np.argmax(predictions[0])
    confidence = predictions[0][predicted_class] * 100

    # Get class name
    class_names = list(train_generator.class_indices.keys())
    predicted_label = class_names[predicted_class]
    predicted_full_name = lesion_type_dict[predicted_label]

    # Create results dictionary
    results = {}
    for idx, class_name in enumerate(class_names):
        full_name = lesion_type_dict[class_name]
        results[full_name] = float(predictions[0][idx])

    # Create detailed output
    output_text = f"""
    ###  Diagnosis: **{predicted_full_name}**

    **Confidence:** {confidence:.2f}%

    ---

    ####  About {predicted_full_name}:
    """

    # Add information about each lesion type
    lesion_info = {
        'Melanocytic nevi': 'Common moles, usually benign. Regular monitoring is recommended.',
        'Melanoma': 'Most serious form of skin cancer. Requires immediate medical attention.',
        'Benign keratosis': 'Non-cancerous skin growth. Generally harmless.',
        'Basal cell carcinoma': 'Most common type of skin cancer. Highly treatable when detected early.',
        'Actinic keratoses': 'Precancerous lesions. Should be monitored and treated.',
        'Vascular lesions': 'Abnormalities in blood vessels. Usually benign.',
        'Dermatofibroma': 'Benign fibrous nodule. Generally harmless.'
    }

    output_text += f"\n{lesion_info[predicted_full_name]}"
    output_text += "\n\n **Disclaimer:** This is an AI model for educational purposes only. Always consult a dermatologist for proper diagnosis."

    return output_text, results

# Create Gradio interface
with gr.Blocks(theme=gr.themes.Soft()) as demo:
    gr.Markdown(
        """
        #  Skin Cancer Classification System
        ### AI-Powered Skin Lesion Analysis

        Upload an image of a skin lesion to get an AI-powered classification. This system can identify 7 different types of skin lesions.

        ** Important:** This tool is for educational purposes only and should not replace professional medical advice.
        """
    )

    with gr.Row():
        with gr.Column():
            input_image = gr.Image(label="Upload Skin Lesion Image")
            predict_btn = gr.Button(" Analyze Lesion", variant="primary", size="lg")

            gr.Markdown(
                """
                ###  Lesion Types:
                - **Melanocytic nevi** - Common moles
                - **Melanoma** - Malignant tumor
                - **Benign keratosis** - Seborrheic keratosis
                - **Basal cell carcinoma** - Common skin cancer
                - **Actinic keratoses** - Precancerous lesions
                - **Vascular lesions** - Blood vessel abnormalities
                - **Dermatofibroma** - Benign fibrous nodule
                """
            )

        with gr.Column():
            output_text = gr.Markdown(label="Diagnosis")
            output_plot = gr.Label(label="Confidence Scores", num_top_classes=7)

    predict_btn.click(
        fn=predict_skin_lesion,
        inputs=input_image,
        outputs=[output_text, output_plot]
    )

    gr.Markdown(
        """
        ---
        ###  Model Information:
        - **Architecture:** EfficientNetB0 with Transfer Learning
        - **Training Dataset:** HAM10000 (10,000+ dermatoscopic images)
        - **Classes:** 7 types of skin lesions
        - **Input Size:** 224x224 pixels

        ###  How to Use:
        1. Upload a clear image of the skin lesion
        2. Click "Analyze Lesion" button
        3. Review the AI prediction and confidence scores
        4. **Always consult a dermatologist for professional diagnosis**
        """
    )

# Step 16: Launch Gradio interface
print("\n" + "="*50)
print("Launching Gradio interface...")
print("="*50)

demo.launch(share=True, debug=True)
```
